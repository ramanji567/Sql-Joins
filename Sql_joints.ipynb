```python
import sqlite3
import pandas as pd

# Using a 'with' statement ensures the connection is properly closed
# even if errors occur, preventing 'database is locked' errors.
conn = sqlite3.connect('sql_joins.db')
cursor = conn.cursor()

    # Drop tables if they exist to ensure a clean slate on re-executi

cursor.execute(
        '''CREATE TABLE IF NOT EXISTS students(
          student_id INTEGER PRIMARY KEY AUTOINCREMENT,
          name TEXT,
          email TEXT,
          City TEXT
        )''')
cursor.execute(
        '''CREATE TABLE IF NOT EXISTS Enrollments(
          enrollment_id INTEGER PRIMARY KEY,
          student_id INTEGER,
          course_name TEXT,
          enrollment_date TEXT,
          FOREIGN KEY (student_id) REFERENCES students(student_id)
        )'''
    )
cursor.execute(
        '''CREATE TABLE IF NOT EXISTS grades(
          grade_id INTEGER PRIMARY KEY,
          student_id INTEGER,
          course_name TEXT,
          score INTEGER,
          FOREIGN KEY (student_id) REFERENCES students(student_id)
        )''')

students_data = [
        ('Alice','alice@email.com','Mumbai'),
        ('Bob','bob@email.com','Delhi'),
        ('Carol','carole@email.com','Bangalore'),
        ('David','david@email.com','Chennai'),
        ('Emma','emma@email.com','Pune')
    ]
cursor.executemany('INSERT INTO students(name,email,city) VALUES(?,?,?)',students_data)

enrollments_data = [
        (101,1,'Python','2024-01-15'),
        (102,1,'SQL','2024-02-01'),
        (103,2,'Data Science','2024-01-20'),
        (104,3,'Python','2024-02-10'),
        (105,3,'Machine Learning','2024-01-25'),
    ]

cursor.executemany('''INSERT INTO enrollments(enrollment_id,student_id,course_name,enrollment_date)
    VALUES(?,?,?,?)''',enrollments_data)

grades_data = [
        (201,1,'Python',85),
        (202,1,'SQL',90),
        (203,2,'Data Science',78),
        (204,4,'Web Development',88)
    ]

cursor.executemany('''INSERT INTO grades(grade_id,student_id,course_name,score)
    VALUES(?,?,?,?)''',grades_data)

    # Commit changes within the 'with' block
conn.commit()

# The connection is automatically closed when exiting the 'with' block
```

Task 1:Basic Joins
    
    1.INNER JOIN

```python
conn = sqlite3.connect('sql_joins.db')
cursor = conn.cursor()
query = '''SELECT s.name,e.course_name
    FROM students  AS s
    INNER JOIN enrollments  AS e ON s.student_id = e.student_id'''
cursor.execute(query)
cursor.fetchall()
```

INNER JOIN means extra the data both tables have common id only extract common data only in both columns

2.Left JOIN

```python
conn = sqlite3.connect('sql_joins.db')
cursor = conn.cursor()
query = '''SELECT s.name ,e.course_name
FROM students AS s
LEFT JOIN enrollments AS e ON s.student_id = e.student_id'''
cursor.execute(query)
cursor.fetchall()
#
```

LEFT JOIN

     In left join to extra the from left table to match with right table fill with right table data if there is no match in right table fill with null values

Task 2 : Multiple Table Joins

```python
conn = sqlite3.connect('sql_joins.db')
cursor = conn.cursor()
query = '''SELECT s.name,e.course_name,g.score
FROM students AS s
LEFT JOIN enrollments AS e ON s.student_id = e.student_id
LEFT JOIN grades AS g ON s.student_id = g.student_id  AND e.course_name = g.course_name'''
cursor.execute(query)
cursor.fetchall()
```

MULTIPLE JOINS

        in this query two left joins are extra the data all from left table match right table when there is no match fill null values in this query

```python
conn = sqlite3.connect('sql_joins.db')
query = '''SELECT * FROM enrollments'''
enrollments_df = pd.read_sql_query(query,conn)
enrollments_df
```

```python
conn = sqlite3.connect('sql_joins.db')
query = '''SELECT * FROM grades'''
grades_df = pd.read_sql(query,conn)
grades_df
```

```python
result = pd.merge(student_df,enrollments_df,on = 'student_id',how ='left' )
result
```

```python
result2 = pd.merge(result,grades_df,on = ['student_id','course_name'],how = 'left')
result2
```
